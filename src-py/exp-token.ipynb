```python
import numpy as np
import matplotlib.pyplot as plt
```

tokenizer profiling

```python
with open("../logs/token-mac.out") as f:
    """
    Elapsed cycles: 14553532
    Elapsed time: 7288 us
    (cycs/token): 4851177.500000
    (us/token): 2429.333252 us
    """
    lines = f.readlines()
    mac_cyc_per_token = np.array([float(l.split()[1]) for l in lines if "cycs/token" in l])
    mac_usec_per_token = np.array([float(l.split()[1]) for l in lines if "us/token" in l])

with open("../logs/token-pi.out") as f:
    """
    encoding time: 1804461550 cycles
    encoding time: 886116181 usec
    encoding time: 886 sec
    cycs/token: 601487183
    usec/token: 295372060
    sec/token: 295
    DONE!!!
    """
    lines = f.readlines()
    pi_cyc_per_token = np.array([float(l.split()[1]) for l in lines if "cycs/token" in l])
    pi_usec_per_token = np.array([int(l.split()[1]) for l in lines if "usec/token" in l])
```

```python
# plot histogram for both and plot means on log scale
# x axis is log cycle, y is frequency
fig, ax = plt.subplots()
ax.hist(np.log(mac_cyc_per_token), bins=50, alpha=0.5, label="Mac")
ax.hist(np.log(pi_cyc_per_token), bins=50, alpha=0.5, label="r/pi")
ax.set_xscale('log')
ax.set_xlabel("log(cycles/token)")
ax.set_ylabel("Frequency")
ax.legend()
plt.show()

fig.clear()
fig, ax = plt.subplots()
ax.hist(np.log(mac_usec_per_token), bins=50, alpha=0.5, label="Mac")
ax.hist(np.log(pi_usec_per_token), bins=50, alpha=0.5, label="r/pi")
ax.set_xscale('log')
ax.set_xlabel("log(usec/token)")
ax.set_ylabel("Frequency")
ax.legend()
plt.show()
```
